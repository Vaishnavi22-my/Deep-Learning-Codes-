```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input1'] = [0,0,1,1]
or_data['input2'] = [0,1,0,1]
or_data['output'] = [0,1,1,1]
```

```python
and_data['input1'] = [0,0,1,1]
and_data['input2'] = [0,1,0,1]
and_data['output'] = [0,0,0,1]
```

```python
xor_data['input1'] = [0,0,1,1]
xor_data['input2'] = [0,1,0,1]
xor_data['output'] = [0,1,1,0]
```

```python
or_data
```

```python
sns.scatterplot(data = or_data , x= 'input1' , y= 'input2' , hue = 'output', s=100)
```

```python
and_data
```

```python
sns.scatterplot(data = and_data , x= 'input1' , y= 'input2' , hue = 'output', s=100)
```

```python
xor_data
```

```python
sns.scatterplot(data = xor_data , x= 'input1' , y= 'input2' , hue = 'output', s=100)
```

```python
from sklearn.linear_model import Perceptron
clf1 = Perceptron()
clf2 = Perceptron()
clf3 = Perceptron()
```

```python
clf1.fit(or_data[['input1' , 'input2']] , or_data['output'])
clf2.fit(and_data[['input1', 'input2']] , and_data['output'])
clf3.fit(xor_data[['input1' , 'input2']] , xor_data['output'])
```

```python
clf1.coef_
```

```python
clf1.intercept_
```

```python
m = -clf1.coef_[0][0]/clf1.coef_[0][1]
c = -clf1.intercept_/clf1.coef_[0][1]
```

```python
x = np.linspace(0,1,5)
y = m*x + c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x , y , color ='red' , linewidth = 3)
sns.scatterplot(data = or_data , x= 'input1' , y= 'input2' , hue = 'output', s=100)
```

```python
m1 = -clf2.coef_[0][0]/clf2.coef_[0][1]
c1 = -clf2.intercept_/clf2.coef_[0][1]
```

```python
x1 = np.linspace(0,1,5)
y1 = m1*x1 + c1
```

```python
plt.figure(figsize=(10,6))
plt.plot(x1 , y1 , color ='red' , linewidth = 3)
sns.scatterplot(data = and_data , x= 'input1' , y= 'input2' , hue = 'output', s=100)
```

```python
m2 = -clf3.coef_[0][0]/clf3.coef_[0][1]
c2 = -clf3.intercept_/clf3.coef_[0][1]
```

```python
x2 = np.linspace(0,1,5)
y2 = m2*x2 + c2

plt.figure(figsize=(10,6))
plt.plot(x2 , y2 , color ='red' , linewidth = 3)
sns.scatterplot(data = xor_data , x= 'input1' , y= 'input2' , hue = 'output', s=100)
```
